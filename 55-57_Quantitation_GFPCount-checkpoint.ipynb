```python
import lumicks.pylake as lk
```

```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#DNA1 Condition1 WITHOUT 55-57 from 11-13
```

```python
file = lk.File("20241113-111644 Kymograph C1_DNA1_Kymo11_13.h5")
```

```python
print(file)
```

```python
print(file.kymos)
```

```python
for name, kymo in file.kymos.items():
    print(f"kymograph '{name}', starts at {kymo.start} ns")
```

```python
rgb = kymo.get_image("rgb")
```

```python
f = plt.figure()
f.set_figwidth(10)
f.set_figheight(8)
kymo_new = kymo["0s":"160s"].plot("rgb", adjustment=lk.ColorAdjustment(0, 99, mode="percentile"))
plt.show()
```

```python
f = plt.figure()
f.set_figwidth(10)
f.set_figheight(8)
kymo_new = kymo["0s":"160s"].plot("b", adjustment=lk.ColorAdjustment(0, 99, mode="percentile"))
plt.show()
```

```python
background = kymo["60s":"65s"].crop_by_distance(2.5, 2.7)
blue_background_per_pixel = np.sum(background.get_image("blue"))#Trying background correction
```

```python
# Calculate background-subtracted photon count
background_subtracted_photon_count = file["Photon count"]["Blue"].downsampled_over(kymo.line_timestamp_ranges(), reduce=np.sum) - blue_background_per_pixel

# Plot the background-subtracted photon count
kymo.crop_by_distance(3.0, 5.5).plot_with_channels(
    [
        file.force2x.downsampled_by(100),
        background_subtracted_photon_count,
    ],
    "gb",
    adjustment=lk.ColorAdjustment(0, 90, "percentile"),
    aspect_ratio=0.2,
    title_vertical=True,
    scale_bar=lk.ScaleBar(10.0, 0.5),
    colors=[[1.0, 0.2, 0.5], "blue"],
    labels=["Force on DNA", "Normalized Blue Photon Count"],
)
#plt.savefig('DNA1_Condition1_GreenCount.png')
```
