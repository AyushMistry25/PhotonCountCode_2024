**This code was used to import, process, and visualize kymographs, as well as computing the total photon count from a channel of interest inside a defined region of interest along the distance axis.** 

The required libraries were installed via Anaconda Prompt and called/imported here in the beginning.

```python
import lumicks.pylake as lk
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Once inside the folder/working directory of interest which contains the .h5 kymograph files, we processed each file individually. Below, we have proceeded with one representative kymograph.

```python
file = lk.File("20241113-182918 Kymograph C2R_DNA7_Kymo11_13.h5")
print(file)
print(file.kymos)
```

The following line of code is passed to define the kymograph file of interest that you want to employ for any subsequent code that you run. 
Ensure that if you import a new kymograph file after processing one, you must pass this line of code again to overwrite the previous working kymograph with the new one.

```python
for name, kymo in file.kymos.items():
    print(f"kymograph '{name}', starts at {kymo.start} ns")
```

```python
rgb = kymo.get_image("rgb")
```

First, we plot any short time window of the kymograph with all the colors on to get a better sense of where the ROI lies. 

```python
f = plt.figure()
f.set_figwidth(10)
f.set_figheight(8)
kymo_new = kymo["60s":"80s"].plot("rgb", adjustment=lk.ColorAdjustment(0, 99, mode="percentile"))
plt.show()
```

Now, we crop the kymograph by distance along the vertical axis to define our region of interest. We want the full time window of the kymograph, therefore we do not process the kymograph any further along the horizontal time axis. 

```python
f = plt.figure()
f.set_figwidth(10)
f.set_figheight(8)
kymo_new = kymo["60s":"80s"].crop_by_distance(2.66, 4.64).plot("rgb", adjustment=lk.ColorAdjustment(0, 99, mode="percentile"))
plt.show()
```

To visually inspect your color channel of interest within the defined region, and also plot a photon count plot as a function of time, we run the following code chunk. The force plot is not of immediate use, but can help detect any disturbance on the beads that might have occured due to collision with debris in the microfluidic channel(s). You can also export the data points for the photon count plot as a .csv file. 
The last two lines of the code chunk can be used to save the data for photon counts as a .csv file and save the combined figure as a .png file. 

```python
blue_counts = file["Photon count"]["Blue"].downsampled_over(kymo.line_timestamp_ranges(), reduce=np.sum)
kymo.crop_by_distance(2.66, 4.64).plot_with_channels(
    [
        file.force2x.downsampled_by(100),
        file["Photon count"]["Blue"].downsampled_over(kymo.crop_by_distance(2.66, 4.64).line_timestamp_ranges(), reduce=np.sum),
    ],
    "b",
    adjustment=lk.ColorAdjustment(5, 98, "percentile"),
    aspect_ratio=0.2,
    title_vertical=True,
    scale_bar=lk.ScaleBar(20, 0.5),
    colors=[[1.0, 0.2, 0.5], "blue"],
    labels=["Force on DNA", "Blue Photon Count"],
)
time_counts =blue_counts.timestamps
time_counts = (time_counts - time_counts[0]) *1e-9
photon_counts=blue_counts.data
counts= np.vstack((time_counts,photon_counts)).T
#np.savetxt('C2_DNA1_11-13_BPC.csv',counts, delimiter=',', header='time counts')
#plt.savefig('DNA1_Condition2_11-13_BPC.png')
```
